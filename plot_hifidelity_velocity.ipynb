```python
import xarray as xr
import numpy as np 
import pylab as plt

u10_hires = xr.open_dataset('output_data/U10m_1.nc')['lo_elev'].values
v10_hires = xr.open_dataset('output_data/V10m_1.nc')['lo_elev'].values

u10_hires = u10_hires[0,:,:]
v10_hires = v10_hires[0,:,:]

lat = xr.open_dataset('output_data/U10m_1.nc')['latitude'].values
lon = xr.open_dataset('output_data/U10m_1.nc')['longitude'].values
```

```python
"""plot spatial maps of high and low resolution wind speed using pcolormesh, lat and lon"""
def plot_wind_speed(u10_hires, v10_hires, lat, lon):
    fig, ax = plt.subplots(1,2, figsize = (10,5))
    ax[0].pcolormesh(lon, lat, u10_hires, cmap = 'Reds')
    ax[0].set_title('High resolution u10')
    ax[1].pcolormesh(lon, lat, v10_hires, cmap = 'Reds')
    ax[1].set_title('High resolution v10')
    # insert colorbar for both plots
    fig.subplots_adjust(right=0.8)
    cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
    fig.colorbar(ax[0].pcolormesh(lon, lat, u10_hires, cmap = 'Reds'), cax=cbar_ax)
    
    plt.show()

plot_wind_speed(u10_hires, v10_hires, lat, lon)
```

```python
print('High resolution U10 range = ', np.min(u10_hires), 'to', np.max(u10_hires))
u10_lowres = xr.open_dataset('data/lofid_u10_monthly.nc')['u10'].values
u10_lowres = u10_lowres[0,:,:]
print('Low resolution U10 range = ', np.min(u10_lowres), 'to', np.max(u10_lowres))

print('High resolution V10 range = ', np.min(v10_hires), 'to', np.max(v10_hires))
v10_lowres = xr.open_dataset('data/lofid_v10_monthly.nc')['v10'].values
v10_lowres = v10_lowres[0,:,:]
print('Low resolution V10 range = ', np.min(v10_lowres), 'to', np.max(v10_lowres))


```
