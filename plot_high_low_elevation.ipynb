```python
import iris
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

```

```python
cube = iris.load_cube('data/hifid_t2m_monthly.nc')
# how many elements in the cube
cube = cube[1,:,:]

"""plot the cube onto a map scaled to the latitiude and longitude of the cube"""
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np

# create a figure
fig = plt.figure(figsize=(10, 10))
# create a map projection
ax = plt.axes(projection=ccrs.PlateCarree())
# add coastlines
ax.coastlines()
# add gridlines
ax.gridlines()
# add the cube data to the map
plt.pcolormesh(cube.coord('longitude').points, cube.coord('latitude').points, cube.data)
# add a colorbar
plt.colorbar()
# show the plot
plt.show()



cube.data.shape


```
